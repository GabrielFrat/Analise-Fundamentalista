```python
dfAtivos = filtro2.reset_index()
dfAtivos = dfAtivos.sort_values(by='dy', ascending=False)
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'level_0': 'Pts dy'})
dfAtivos.drop
print(dfAtivos)
```

```python
# Criar pontuador para as melhores empresas
dfAtivos = dfAtivos.drop(columns=['index'])

dfAtivos = dfAtivos.sort_values(by='roe', ascending=False)
dfAtivos = dfAtivos.reset_index()

dfAtivos = dfAtivos.drop(columns=['index'])
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'index': 'Pts roe'})
dfAtivos
```

```python
dfAtivos = dfAtivos.sort_values(by='pl', ascending=False)
dfAtivos = dfAtivos.reset_index()

dfAtivos = dfAtivos.drop(columns=['index'])
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'index': 'Pts pl'})
dfAtivos
```

```python
dfAtivos = dfAtivos.sort_values(by='mrgliq', ascending=False)
dfAtivos = dfAtivos.reset_index()

dfAtivos = dfAtivos.drop(columns=['index'])
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'index': 'Pts Mrg Liq'})
dfAtivos
```

```python
dfAtivos = dfAtivos.sort_values(by='c5y', ascending=False)
dfAtivos = dfAtivos.reset_index()

dfAtivos = dfAtivos.drop(columns=['index'])
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'index': 'Pts c5y'})
dfAtivos
```

```python
dfAtivos = dfAtivos.sort_values(by='liq2m', ascending=False)
dfAtivos = dfAtivos.reset_index()

dfAtivos = dfAtivos.drop(columns=['index'])
dfAtivos = dfAtivos.reset_index()
dfAtivos = dfAtivos.rename(columns={'index': 'Pts liq2m'})
dfAtivos
```

```python
dfAtivos['Pontu√ß√£o'] = dfAtivos['Pts pl'] + dfAtivos['Pts roe'] + dfAtivos['Pts dy'] + dfAtivos['Pts c5y'] + dfAtivos['Pts Mrg Liq'] + dfAtivos['Pts liq2m']
dfAtivos
```

```python
dfAtivos = dfAtivos.sort_values(by='Pontu√ß√£o', ascending=True)
dfAtivos
```

```python
plt.figure(figsize = (22, 10))
x_labels = dfAtivos['Pontu√ß√£o'].tolist()

plt.bar(dfAtivos.papel, dfAtivos.Pontu√ß√£o)
plt.xticks(rotation=45)

plt.title("Lista de Pontu√ß√£o Acumulada")
```
